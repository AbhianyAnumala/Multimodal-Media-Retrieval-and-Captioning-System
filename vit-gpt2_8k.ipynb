```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split, Subset
from torchvision import transforms
from PIL import Image
import os
import collections
import numpy as np
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import matplotlib.pyplot as plt
import random
import pandas as pd
import datetime
import tqdm # For progress bars
import gc # Import garbage collector

# New imports for ViT and GPT-2 from Hugging Face Transformers
from transformers import ViTFeatureExtractor, ViTModel, GPT2Tokenizer, GPT2LMHeadModel
from transformers.optimization import get_scheduler
from torch.optim import AdamW # Explicitly import AdamW from torch.optim

# Import for Mixed Precision Training
from torch.cuda.amp import autocast, GradScaler

# Imports for COCO-Eval metrics (for validation loop)
from pycocoevalcap.bleu.bleu import Bleu
# from pycocoevalcap.meteor.meteor import Meteor # Meteor requires Java, might be tricky to set up. Keeping it commented for simplicity.
import json # Required by pycocoevalcap for dictionary formatting

# Import for ROUGE metric
from rouge_score import rouge_scorer
from sklearn.model_selection import train_test_split # For creating random subsets

# Ensure nltk punkt tokenizer data is available for BLEU score calculation.
# Downloads it if not already present.
nltk.download('punkt')

# Clear CUDA cache at the start
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()
    print("CUDA cache cleared.")

"""import kagglehub

# Download latest version
path = kagglehub.dataset_download("adityajn105/flickr8k")

print("Path to dataset files:", path)

import os
!mkdir -p /workspace/flickr8k
!mv $path /workspace/flickr8k
path= "/workspace/flickr8k/1"

# Step 4: Set paths

# Path to the dataset directory
# Path to the dataset directory
dataset_path = path
image_folder = os.path.join(dataset_path, "Images")
caption_file = os.path.join(dataset_path, "captions.txt")
"""
!mkdir -p /workspace/models


# --- 0. Configuration and Hyperparameters ---
CONFIG = {
    "image_dir": "/workspace/flickr8k/1/Images", # Directory where image files are stored
    "captions_path": "/workspace/flickr8k/1/captions.txt", # Path to the captions file
    "vocab_threshold": 1,           # Minimum word count to be included in the vocabulary.
                                    # For GPT-2 tokenizer, this is less critical as it uses subwords.
    "max_seq_length": 20,           # Maximum length of a caption sequence (including special tokens like SOS, EOS, PAD).
                                    # Captions longer than this will be truncated, shorter will be padded.
    "batch_size": 128,              # Number of samples (image, caption pairs) processed in one forward/backward pass.
    "embed_size": 524,              # Dimensionality of the image features projected from ViT.
    "hidden_size": 768,             # GPT-2 small's default hidden size. This is fixed by the pre-trained GPT-2 model.
    "num_layers": 1,                # Not directly used by GPT-2 (it has its own layers), but kept for consistency.
    "learning_rate": 5e-5,          # Recommended learning rate for fine-tuning pre-trained Transformers.
    "num_epochs": 25,               # Number of times to iterate over the entire training dataset.
    "log_step": 100,                # Print training progress every N batches.
    "save_step": 1,                 # Save model checkpoints every N epochs.
    "model_path": "/workspace/models/", # Directory to save trained model weights.
    "vocab_path": "vocab.pkl",      # Path to save/load the custom vocabulary object (less critical with HF tokenizer).\
    "test_predictions_file": "/workspace/TestPredictions_VIT_GPT2.csv", # Output file for test predictions.
    "test_scores_file": "/workspace/TestScores_VIT_GPT2.csv",           # Output file for evaluation scores.
    "vit_model_name": "google/vit-base-patch16-224-in21k", # Pre-trained Vision Transformer model name.
    "gpt2_model_name": "gpt2",      # Pre-trained GPT-2 language model name.
    "beam_size": 4,                 # Number of beams for beam search decoding during inference.
    "val_subset_ratio": 0.2,        # Percentage of validation data to use for quick validation (e.g., 0.2 = 20%)
    "calculate_individual_bleu_for_histograms": True # Set to False for faster evaluation if histograms aren't needed
}

# Initialize a smoothing function for BLEU score calculation to handle zero matches.
smooth_fn = SmoothingFunction().method1

# --- 1. Vocabulary Class (Adapted for GPT-2 Tokenizer) ---
class Vocabulary(object):
    """
    Custom Vocabulary class that wraps Hugging Face's GPT2Tokenizer.
    This class provides a consistent interface for the Dataset,
    while leveraging the advanced tokenization capabilities of GPT-2.
    """
    def __init__(self, vocab_threshold, vocab_file='vocab.pkl', gpt2_model_name="gpt2"):
        self.vocab_threshold = vocab_threshold
        self.vocab_file = vocab_file
        self.tokenizer = GPT2Tokenizer.from_pretrained(gpt2_model_name)

        # Add special tokens if they are not already in GPT-2 tokenizer's vocabulary.
        if self.tokenizer.pad_token is None:
            self.tokenizer.add_special_tokens({'pad_token': '<PAD>'})
        if self.tokenizer.bos_token is None:
            self.tokenizer.add_special_tokens({'bos_token': '<SOS>'})
        if self.tokenizer.eos_token is None:
            self.tokenizer.add_special_tokens({'eos_token': '<EOS>'})
        
        # Re-initialize word2idx and idx2word to reflect any newly added special tokens.
        self.word2idx = self.tokenizer.get_vocab()
        self.idx2word = {idx: word for word, idx in self.word2idx.items()}
        self.create_vocab = True # Vocabulary is effectively built from the tokenizer.

    def build_vocabulary(self, captions):
        """
        With GPT-2 tokenizer, the vocabulary is pre-built. This method is a placeholder
        and does not re-build the vocabulary. It just prints a message.
        """
        print("Using pre-trained GPT-2 tokenizer vocabulary. `build_vocabulary` does not re-build.")
        pass

    def __call__(self, word):
        """
        Allows direct lookup of a word's token ID. For single words, this is a simplification.
        For full sentences, `tokenizer.encode()` should be used.
        """
        token_id = self.tokenizer.convert_tokens_to_ids(word)
        if token_id is not None:
            return token_id
        else:
            print(f"Warning: '{word}' not directly found as a single token. Returning pad_token_id. "
                  f"For full sentence encoding, use tokenizer.encode().")
            return self.tokenizer.pad_token_id

    def __len__(self):
        """Returns the size of the tokenizer's vocabulary."""
        return len(self.tokenizer)

# --- 2. Dataset Class (Adapted for ViT and GPT-2 Tokenizer) ---
class Flickr8kDataset(Dataset):
    """
    Custom Dataset class for the Flickr8k dataset, adapted for ViT image processing
    and GPT-2 text tokenization.
    """
    def __init__(self, image_dir, captions_path, vocab, transform=None, max_seq_length=20, gpt2_model_name="gpt2"):
        self.image_dir = image_dir
        self.captions = self._load_captions(captions_path) # Load image-caption mappings
        self.image_names = list(self.captions.keys()) # List of unique image filenames
        self.vocab = vocab # The Vocabulary object (wrapping GPT2Tokenizer)
        self.transform = transform # Image transformations (e.g., resize, normalize)
        self.max_seq_length = max_seq_length # Max length for tokenized captions
        # self.tokenizer = GPT2Tokenizer.from_pretrained(gpt2_model_name) # Redundant, use self.vocab.tokenizer
        self.tokenizer = self.vocab.tokenizer # Use the tokenizer from the vocabulary

        # Ensure special tokens are set for the tokenizer, consistent with Vocabulary class.
        if self.tokenizer.pad_token is None:
            self.tokenizer.add_special_tokens({'pad_token': '<PAD>'})
        if self.tokenizer.bos_token is None:
            self.tokenizer.add_special_tokens({'bos_token': '<SOS>'})
        if self.tokenizer.eos_token is None:
            special_tokens_dict = {'eos_token': '<EOS>'}
            self.tokenizer.add_special_tokens(special_tokens_dict)

        # Create a list of (image_name, caption_text) pairs for training.
        # For evaluation, we will retrieve all captions for an image.
        self.data_pairs = []
        for img_name, caption_list in self.captions.items():
            for caption_text in caption_list:
                self.data_pairs.append((img_name, caption_text))

    def _load_captions(self, captions_path):
        """
        Loads captions from the specified file. If the file is not found,
        it creates dummy images and a dummy captions file for demonstration.
        """
        captions_data = collections.defaultdict(list)
        try:
            with open(captions_path, 'r', encoding='utf-8') as f:
                next(f) # Skip header line
                for line in f:
                    line = line.strip()
                    if not line: # Skip empty lines
                        continue
                    parts = line.split(',', 1) # Split only on the first comma
                    if len(parts) != 2:
                        print(f"Skipping malformed line: {line} (Expected 'image,caption' format)")
                        continue
                    image_filename, caption = parts[0], parts[1]
                    # Remove the '#id' suffix from the image filename (e.g., 'image.jpg#0' -> 'image.jpg')
                    image_filename = image_filename.split('#')[0].strip()
                    captions_data[image_filename].append(caption.strip())
            print(f"Loaded {sum(len(v) for v in captions_data.values())} captions for {len(captions_data)} images from {captions_path}")
            return dict(captions_data)
        except FileNotFoundError:
            print(f"Caption file '{captions_path}' not found. Please ensure it's in the correct directory.")
            print("Creating dummy images and a dummy .txt caption file for demonstration.")
            return self._create_dummy_captions()

    def _create_dummy_captions(self, num_dummy_images=100):
        """
        Creates dummy image files and a dummy captions.txt file for testing
        if the actual dataset is not available.
        """
        dummy_captions = collections.defaultdict(list)
        os.makedirs(CONFIG["image_dir"], exist_ok=True) # Ensure image directory exists

        lines_to_write = []
        lines_to_write.append("image_name#id,caption_text") # Header for the dummy file

        for i in range(num_dummy_images):
            img_name = f"dummy_image_{i:04d}.jpg" # Formatted image name
            # Create a dummy image (224x224, random color)
            Image.new('RGB', (224, 224), color = (random.randint(0,255), random.randint(0,255), random.randint(0,255))).save(os.path.join(CONFIG["image_dir"], img_name))

            # Generate 5 dummy captions for each image
            caption1 = f"a dummy caption for image number {i}."
            caption2 = f"this is another example caption for dummy image {i}, with a comma."
            caption3 = f"dummy image {i} shows a test scene with a colorful background."
            caption4 = f"a randomly generated colored box labeled as {i}."
            caption5 = f"simple abstract art with the number {i} in the center."

            dummy_captions[img_name].extend([caption1, caption2, caption3, caption4, caption5])

            # Add to lines to write with the Flickr30k format (image_name#id,caption)
            lines_to_write.append(f"{img_name}#0,{caption1}")
            lines_to_write.append(f"{img_name}#1,{caption2}")
            lines_to_write.append(f"{img_name}#2,{caption3}")
            lines_to_write.append(f"{img_name}#3,{caption4}")
            lines_to_write.append(f"{img_name}#4,{caption5}")

        with open(CONFIG["captions_path"], 'w', encoding='utf-8') as f:
            for line in lines_to_write:
                f.write(line + '\n')

        print(f"Created {num_dummy_images} dummy images in {CONFIG['image_dir']}")
        print(f"Dummy .txt captions created at {CONFIG['captions_path']}")
        return dict(dummy_captions)

    def __len__(self):
        """Returns the total number of (image, caption) pairs."""
        return len(self.data_pairs)

    def __getitem__(self, idx):
        """
        Retrieves an image and its corresponding tokenized caption at a given index.
        Returns all ground truth captions for the image, and a single caption for training.
        """
        img_name, training_caption_text = self.data_pairs[idx]
        image_path = os.path.join(self.image_dir, img_name)
        image = Image.open(image_path).convert('RGB') # Load image and convert to RGB

        if self.transform:
            image = self.transform(image) # Apply image transformations

        # Tokenize caption using GPT-2 tokenizer.
        encoded_caption = self.tokenizer.encode(
            self.tokenizer.bos_token + training_caption_text + self.tokenizer.eos_token,
            max_length=self.max_seq_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        )
        caption_indices = encoded_caption.squeeze(0)

        # Get ALL ground truth captions for this specific image for evaluation metrics
        all_ground_truth_captions_for_image = self.captions.get(img_name, [])

        # Return the processed image, tokenized caption (for training), original image name,
        # and ALL raw caption texts for evaluation.
        return image, caption_indices, img_name, all_ground_truth_captions_for_image

# --- 3. Encoder (Vision Transformer - ViT) ---
class EncoderViT(nn.Module):
    """
    Encoder module using a pre-trained Vision Transformer (ViT) to extract image features.
    """
    def __init__(self, embed_size, vit_model_name=CONFIG["vit_model_name"]):
        super(EncoderViT, self).__init__()
        self.vit = ViTModel.from_pretrained(vit_model_name)
        vit_hidden_size = self.vit.config.hidden_size
        self.linear = nn.Linear(vit_hidden_size, embed_size)
        self.bn = nn.BatchNorm1d(embed_size, momentum=0.01)

    def forward(self, images):
        """
        Forward pass for the EncoderViT.
        """
        outputs = self.vit(images)
        features = outputs.last_hidden_state[:, 0, :] # [CLS] token embedding
        features = self.bn(self.linear(features))
        return features

# --- 4. Decoder (GPT-2 Language Model) ---
class DecoderGPT2(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size, gpt2_model_name="gpt2"):
        super(DecoderGPT2, self).__init__()
        self.gpt2 = GPT2LMHeadModel.from_pretrained(gpt2_model_name)
        self.tokenizer = GPT2Tokenizer.from_pretrained(gpt2_model_name)

        # Add special tokens if not already present, including a placeholder for the image
        special_tokens_dict = {}
        if self.tokenizer.pad_token is None:
            special_tokens_dict["pad_token"] = '<PAD>'
        if self.tokenizer.bos_token is None:
            special_tokens_dict["bos_token"] = '<SOS>'
        if self.tokenizer.eos_token is None:
            special_tokens_dict["eos_token"] = '<EOS>'
        
        if self.tokenizer.convert_tokens_to_ids('<IMG>') is None:
            special_tokens_dict["additional_special_tokens"] = ['<IMG>']

        if special_tokens_dict:
            num_added_toks = self.tokenizer.add_special_tokens(special_tokens_dict)
            self.gpt2.resize_token_embeddings(len(self.tokenizer))
            print(f"Added {num_added_toks} new tokens to GPT-2 tokenizer and resized embeddings.")

        self.image_token_id = self.tokenizer.convert_tokens_to_ids('<IMG>')
        if self.image_token_id is None:
             print("Warning: <IMG> token ID not found, potentially affecting generation conditioning.")

        # Linear layer to project image features to the GPT-2 hidden size
        self.image_feature_projection = nn.Linear(embed_size, hidden_size)

    def forward(self, features, captions):
        # features: (batch_size, embed_size) - image features from encoder
        # captions: (batch_size, max_seq_length) - tokenized captions

        projected_features = self.image_feature_projection(features) # (batch_size, hidden_size)

        # Create input_ids that start with <IMG> token, followed by actual caption tokens (including <SOS>)
        img_token_ids_tensor = torch.full((captions.shape[0], 1), self.image_token_id, device=captions.device, dtype=torch.long)
        combined_input_ids = torch.cat((img_token_ids_tensor, captions), dim=1) # (batch_size, 1 + max_seq_length)

        # Get embeddings for the combined input IDs from GPT-2's embedding layer
        inputs_embeds = self.gpt2.transformer.wte(combined_input_ids)

        # Overwrite the embedding for the <IMG> token (at index 0) with the projected image features
        inputs_embeds[:, 0, :] = projected_features

        # Create attention mask, including the image token.
        attention_mask = (combined_input_ids != self.tokenizer.pad_token_id).long()

        # Prepare labels: shift original captions by one, ignore image token label.
        labels = combined_input_ids.clone()
        labels[:, 0] = -100 # Ignore loss for the image token (first position)
        
        outputs = self.gpt2(inputs_embeds=inputs_embeds, attention_mask=attention_mask, labels=labels)
        
        # Return both logits and loss
        return outputs.logits, outputs.loss

    def generate(self, image_features, vocab, device, max_length=20, beam_size=4):
        # image_features: (batch_size, embed_size) - batch of image features
        
        self.eval() # Ensure model is in evaluation mode for inference

        # Project image features to GPT-2's hidden size
        projected_features = self.image_feature_projection(image_features) # (batch_size, hidden_size)

        # Get the embedding for the <SOS> token
        # Expand sos_token_embedding to match the batch_size of projected_features
        sos_token_embedding_repeated = self.gpt2.transformer.wte(
            torch.tensor([vocab.tokenizer.bos_token_id], device=device)
        ).unsqueeze(0).expand(image_features.size(0), 1, -1) # (batch_size, 1, hidden_size)
        
        # Concatenate projected image features and SOS embeddings
        # Unsqueeze projected_features to (batch_size, 1, hidden_size) to match sos_token_embedding_repeated
        initial_inputs_embeds = torch.cat((projected_features.unsqueeze(1), sos_token_embedding_repeated), dim=1) # (batch_size, 2, hidden_size)
        
        # Create an attention mask for the initial embeddings (image + SOS)
        attention_mask = torch.ones(initial_inputs_embeds.shape[0], initial_inputs_embeds.shape[1], device=device, dtype=torch.long)

        # Generate caption using GPT-2's `generate` method
        generated_ids = self.gpt2.generate(
            inputs_embeds=initial_inputs_embeds, # Pass pre-computed embeddings directly
            attention_mask=attention_mask,
            max_new_tokens=max_length - initial_inputs_embeds.shape[1], # Max new tokens to generate after initial tokens
            num_beams=beam_size,
            early_stopping=True, # Stop when all beams have finished
            eos_token_id=vocab.tokenizer.eos_token_id,
            pad_token_id=vocab.tokenizer.pad_token_id,
            no_repeat_ngram_size=2, # Prevent repeating ngrams for better quality
            length_penalty=0.7, # Encourage shorter sequences
            do_sample=False, # Use beam search (deterministic)
        )
        
        # Decode the generated IDs. The `generate` method returns the full sequence
        # including the initial `inputs_embeds` if they were provided.
        # So, we need to skip the number of tokens that were part of `initial_inputs_embeds` (which is 2: image + SOS)
        decoded_captions = []
        for i in range(generated_ids.size(0)):
            decoded_caption = vocab.tokenizer.decode(
                generated_ids[i][initial_inputs_embeds.shape[1]:], # Slice to remove initial tokens
                skip_special_tokens=True # Skip special tokens like <EOS>, <PAD>
            )
            decoded_captions.append(decoded_caption.strip())
        
        return decoded_captions # Return a list of decoded captions

# --- 5. Combined Image Captioning Model (ViT + GPT-2) ---
class ImageCaptioningModel_ViT_GPT2(nn.Module):
    """
    Combines the ViT Encoder and GPT-2 Decoder into a single image captioning model.
    """
    def __init__(self, embed_size, hidden_size, vocab_size, vit_model_name, gpt2_model_name):
        super(ImageCaptioningModel_ViT_GPT2, self).__init__()
        self.encoder = EncoderViT(embed_size, vit_model_name)
        self.decoder = DecoderGPT2(embed_size, hidden_size, vocab_size, gpt2_model_name)

    def forward(self, images, captions):
        """
        Forward pass for the combined model during training.
        """
        image_features = self.encoder(images)
        logits, loss = self.decoder(image_features, captions)
        return logits, loss

    def generate_caption(self, image_batch, vocab, device, max_length=CONFIG["max_seq_length"], beam_size=CONFIG["beam_size"]):
        """
        Generates captions for a batch of images.
        """
        self.eval() # Set model to evaluation mode
        with torch.no_grad():
            image_features = self.encoder(image_batch) # (batch_size, embed_size)
            generated_captions = self.decoder.generate(image_features, vocab, device,
                                                       max_length=max_length, beam_size=beam_size)
            return generated_captions

# --- Helper Function for ROUGE calculation ---
def calculate_rouge_scores(predictions, references_list_of_lists):
    """
    Calculates ROUGE-1, ROUGE-2, and ROUGE-L scores by averaging scores
    against multiple references for each prediction.
    Args:
        predictions (list of str): List of generated caption strings.
        references_list_of_lists (list of list of str): List where each inner list contains
                                                        all ground truth caption strings for a sample.
    Returns:
        dict: Aggregated ROUGE scores.
    """
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    
    # Initialize lists to store scores for each sample, across all references
    all_sample_rouge_scores = collections.defaultdict(lambda: collections.defaultdict(list))
    
    total_samples = len(predictions)
    if total_samples == 0:
        return {
            'rouge1': {'precision': 0, 'recall': 0, 'fmeasure': 0},
            'rouge2': {'precision': 0, 'recall': 0, 'fmeasure': 0},
            'rougeL': {'precision': 0, 'recall': 0, 'fmeasure': 0},
        }

    for i in range(total_samples):
        current_references = references_list_of_lists[i] # A list of strings for multiple references
        current_prediction = predictions[i] # A single string

        # Calculate ROUGE for each reference and store the best score (max F-measure)
        # or average them. For now, let's average.
        
        # Initialize scores for the current sample
        sample_scores = collections.defaultdict(lambda: {'precision': 0, 'recall': 0, 'fmeasure': 0})
        num_references = len(current_references)
        
        if num_references == 0:
            # If no references, skip this sample or assign zero scores
            continue

        for ref_text in current_references:
            # Each call to scorer.score takes one reference string and one prediction string
            scores_for_ref = scorer.score(ref_text, current_prediction)
            
            for rouge_type in ['rouge1', 'rouge2', 'rougeL']:
                sample_scores[rouge_type]['precision'] += scores_for_ref[rouge_type].precision
                sample_scores[rouge_type]['recall'] += scores_for_ref[rouge_type].recall
                sample_scores[rouge_type]['fmeasure'] += scores_for_ref[rouge_type].fmeasure
        
        # Average the scores for this sample across its references
        for rouge_type in ['rouge1', 'rouge2', 'rougeL']:
            all_sample_rouge_scores[rouge_type]['precision'].append(sample_scores[rouge_type]['precision'] / num_references)
            all_sample_rouge_scores[rouge_type]['recall'].append(sample_scores[rouge_type]['recall'] / num_references)
            all_sample_rouge_scores[rouge_type]['fmeasure'].append(sample_scores[rouge_type]['fmeasure'] / num_references)

    # Aggregate (average) scores across all samples
    final_aggregated_scores = {
        'rouge1': {'precision': 0, 'recall': 0, 'fmeasure': 0},
        'rouge2': {'precision': 0, 'recall': 0, 'fmeasure': 0},
        'rougeL': {'precision': 0, 'recall': 0, 'fmeasure': 0},
    }

    if total_samples > 0:
        for rouge_type in ['rouge1', 'rouge2', 'rougeL']:
            if all_sample_rouge_scores[rouge_type]['precision']: # Check if list is not empty
                final_aggregated_scores[rouge_type]['precision'] = np.mean(all_sample_rouge_scores[rouge_type]['precision'])
                final_aggregated_scores[rouge_type]['recall'] = np.mean(all_sample_rouge_scores[rouge_type]['recall'])
                final_aggregated_scores[rouge_type]['fmeasure'] = np.mean(all_sample_rouge_scores[rouge_type]['fmeasure'])
    
    return final_aggregated_scores

# --- Helper Function for Metric Calculation during Validation ---
def calculate_validation_metrics(model, data_loader, vocab, device, CONFIG):
    """
    Calculates BLEU (1-4), METEOR, and ROUGE scores for a given model and data loader.
    This function is intended for use during validation, using mixed precision.
    Args:
        model (ImageCaptioningModel_ViT_GPT2): The trained model.
        data_loader (DataLoader): DataLoader for the validation set (can be a subset).
        vocab (Vocabulary): The vocabulary object.
        device (torch.device): The device (CPU/GPU).
        CONFIG (dict): Configuration dictionary.
    Returns:
        dict: A dictionary containing 'Bleu_1', 'Bleu_2', 'Bleu_3', 'Bleu_4', 'METEOR',
              ROUGE scores, and 'loss' scores.
    """
    model.eval() # Set model to evaluation mode
    
    # For pycocoevalcap
    gts = {} # Ground truths for COCO-Eval format
    res = {} # Results (predictions) for COCO-Eval format

    # For rouge_score and NLTK BLEU
    all_predictions_text = []
    all_references_text = [] # List of lists of strings

    total_val_loss = 0.0
    num_batches = 0

    criterion = nn.CrossEntropyLoss(ignore_index=vocab.tokenizer.pad_token_id)

    with torch.no_grad():
        for i, (images, captions_input_ids, img_names, all_raw_captions_list) in enumerate(tqdm.tqdm(data_loader, desc="Validation")):
            images = images.to(device)
            captions_input_ids = captions_input_ids.to(device)

            # --- Mixed Precision Inference for validation loss and generation ---
            with autocast():
                logits, loss = model(images, captions_input_ids)
            total_val_loss += loss.item()
            num_batches += 1
            
            # Generate captions for the entire batch
            predicted_captions_batch = model.generate_caption(images, CONFIG["vocab"], CONFIG["device"],
                                                              max_length=CONFIG["max_seq_length"],
                                                              beam_size=CONFIG["beam_size"])
            
            for j in range(images.size(0)):
                img_name = img_names[j]
                predicted_caption = predicted_captions_batch[j]
                
                # Prepare for COCO-Eval format (corrected to pass list of strings directly)
                image_id = str(img_name)
                gts[image_id] = all_raw_captions_list[j] # Pass the list of strings directly
                res[image_id] = [predicted_caption] # Pass a list containing the single predicted string

                # Collect for ROUGE and NLTK BLEU
                all_predictions_text.append(predicted_caption)
                all_references_text.append(all_raw_captions_list[j]) # This is already a list of strings

    # Calculate scores using pycocoevalcap (BLEU)
    # METEOR is commented out as it requires Java setup
    scorers = [
        (Bleu(4), ["Bleu_1", "Bleu_2", "Bleu_3", "Bleu_4"]),
        # (Meteor(), "METEOR") # Uncomment if Meteor is set up
    ]

    metrics = {}
    for scorer, method in scorers:
        score, _ = scorer.compute_score(gts, res)
        if isinstance(method, list):
            for m, s in zip(method, score):
                metrics[m] = s
        else:
            metrics[method] = score
    
    # Calculate ROUGE scores
    rouge_scores = calculate_rouge_scores(all_predictions_text, all_references_text)
    metrics['ROUGE-1_P'] = rouge_scores['rouge1']['precision']
    metrics['ROUGE-1_R'] = rouge_scores['rouge1']['recall']
    metrics['ROUGE-1_F'] = rouge_scores['rouge1']['fmeasure']
    metrics['ROUGE-2_P'] = rouge_scores['rouge2']['precision']
    metrics['ROUGE-2_R'] = rouge_scores['rouge2']['recall']
    metrics['ROUGE-2_F'] = rouge_scores['rouge2']['fmeasure']
    metrics['ROUGE-L_P'] = rouge_scores['rougeL']['precision']
    metrics['ROUGE-L_R'] = rouge_scores['rougeL']['recall']
    metrics['ROUGE-L_F'] = rouge_scores['rougeL']['fmeasure']

    metrics['loss'] = total_val_loss / num_batches
    return metrics

# --- 6. Training Function ---
def train_model(train_loader, val_loader, vocab, device, CONFIG):
    """
    Trains the image captioning model.
    """
    model = ImageCaptioningModel_ViT_GPT2(CONFIG["embed_size"], CONFIG["hidden_size"], len(vocab),
                                          CONFIG["vit_model_name"], CONFIG["gpt2_model_name"]).to(device)
    
    optimizer = AdamW(model.parameters(), lr=CONFIG["learning_rate"])

    num_training_steps = CONFIG["num_epochs"] * len(train_loader)
    lr_scheduler = get_scheduler(
        "linear",
        optimizer=optimizer,
        num_warmup_steps=0,
        num_training_steps=num_training_steps,
    )

    scaler = GradScaler()

    model.train()
    total_step = 0
    start_time = datetime.datetime.now()

    train_losses = []
    val_losses = []
    val_bleu1_scores = []
    val_bleu2_scores = []
    val_bleu3_scores = []
    val_bleu4_scores = []
    val_meteor_scores = [] # Will be empty if Meteor is not used
    val_rouge1_f_scores = []
    val_rougeL_f_scores = []


    print("\n--- Starting Training ---")
    for epoch in range(CONFIG["num_epochs"]):
        running_train_loss = 0.0
        for i, (images, captions_input_ids, img_names, raw_captions) in enumerate(tqdm.tqdm(train_loader, desc=f"Epoch {epoch+1} Training")):
            images = images.to(device)
            captions_input_ids = captions_input_ids.to(device)

            optimizer.zero_grad()

            with autocast():
                logits, loss = model(images, captions_input_ids)
            
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(optimizer)
            scaler.update()

            lr_scheduler.step()

            total_step += 1
            running_train_loss += loss.item()

            if (i+1) % CONFIG["log_step"] == 0:
                print(f"Epoch [{epoch+1}/{CONFIG['num_epochs']}], Step [{i+1}/{len(train_loader)}], "
                      f"Loss: {loss.item():.4f}, Perplexity: {np.exp(loss.item()):.4f}\n") # Added newline for better readability
                
        avg_train_loss = running_train_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        print(f"Epoch {epoch+1} finished. Average Training Loss: {avg_train_loss:.4f}")

        # --- Validation Phase ---
        print(f"--- Epoch {epoch+1} Validation ---")
        val_metrics = calculate_validation_metrics(model, val_loader, CONFIG["vocab"], CONFIG["device"], CONFIG)
        model.train() # Switch back to train mode

        val_losses.append(val_metrics['loss'])
        val_bleu1_scores.append(val_metrics['Bleu_1'])
        val_bleu2_scores.append(val_metrics['Bleu_2'])
        val_bleu3_scores.append(val_metrics['Bleu_3'])
        val_bleu4_scores.append(val_metrics['Bleu_4'])
        # val_meteor_scores.append(val_metrics.get('METEOR', 0.0)) # Use .get() for optional metrics
        val_rouge1_f_scores.append(val_metrics['ROUGE-1_F'])
        val_rougeL_f_scores.append(val_metrics['ROUGE-L_F'])


        print(f"Epoch [{epoch+1}/{CONFIG['num_epochs']}], Validation Loss: {val_metrics['loss']:.4f}, "
              f"BLEU-1: {val_metrics['Bleu_1']:.4f}, BLEU-4: {val_metrics['Bleu_4']:.4f}, "
              f"ROUGE-1 F: {val_metrics['ROUGE-1_F']:.4f}, ROUGE-L F: {val_metrics['ROUGE-L_F']:.4f}")
        
        # Save the model checkpoints
        if (epoch+1) % CONFIG["save_step"] == 0:
            os.makedirs(CONFIG["model_path"], exist_ok=True)
            torch.save(model.state_dict(), os.path.join(CONFIG["model_path"], f'vit_gpt2_model-{epoch+1}.pth'))
            print(f"Model saved to {os.path.join(CONFIG['model_path'], f'vit_gpt2_model-{epoch+1}.pth')}")

    end_time = datetime.datetime.now()
    print(f"\n--- Training Complete in {end_time - start_time} ---")
    return model, train_losses, val_losses, val_bleu1_scores, val_bleu2_scores, val_bleu3_scores, val_bleu4_scores, val_meteor_scores, val_rouge1_f_scores, val_rougeL_f_scores

# --- 7. Evaluation Function (for Test Set) ---
def evaluate_model(model, data_loader, vocab, device, CONFIG):
    """
    Evaluates the trained image captioning model on a test dataset.
    Calculates BLEU (1-4) and ROUGE (1, 2, L) scores.
    """
    model.eval() # Set model to evaluation mode
    predictions_df_data = [] # To store detailed prediction results
    
    gts = {} # Ground truths for COCO-Eval format
    res = {} # Results (predictions) for COCO-Eval format

    all_predictions_text = []
    all_references_text = [] # List of lists of strings

    # Lists to store individual BLEU scores for histogram plotting
    individual_bleu1_scores = []
    individual_bleu2_scores = []
    individual_bleu3_scores = []
    individual_bleu4_scores = []

    print("\n--- Starting Evaluation on Test Set ---")
    with torch.no_grad(): # Disable gradient calculation for evaluation
        for i, (images, _, img_names, all_raw_captions_list) in enumerate(tqdm.tqdm(data_loader, desc="Evaluating Test Set")):
            images = images.to(device)
            
            # Initialize predicted_captions_batch to an empty list before generation
            predicted_captions_batch = [] 
            
            if images.size(0) > 0: # Only generate if there are images in the batch
                # Generate captions for the entire batch
                predicted_captions_batch = model.generate_caption(images, vocab, device,
                                                               max_length=CONFIG["max_seq_length"],
                                                               beam_size=CONFIG["beam_size"])

            for j in range(images.size(0)):
                img_name = img_names[j]
                predicted_caption = predicted_captions_batch[j]
                
                ground_truth_captions_for_image = all_raw_captions_list[j] # This is already a list of strings

                if CONFIG["calculate_individual_bleu_for_histograms"]: # Conditional execution
                    # Calculate individual BLEU scores for this sample (for plotting histograms)
                    references_for_bleu = [nltk.word_tokenize(cap.lower()) for cap in ground_truth_captions_for_image]
                    candidate_for_bleu = nltk.word_tokenize(predicted_caption.lower())

                    bleu1 = sentence_bleu(references_for_bleu, candidate_for_bleu, weights=(1, 0, 0, 0), smoothing_function=smooth_fn)
                    bleu2 = sentence_bleu(references_for_bleu, candidate_for_bleu, weights=(0.5, 0.5, 0, 0), smoothing_function=smooth_fn)
                    bleu3 = sentence_bleu(references_for_bleu, candidate_for_bleu, weights=(0.33, 0.33, 0.33, 0), smoothing_function=smooth_fn)
                    bleu4 = sentence_bleu(references_for_bleu, candidate_for_bleu, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smooth_fn)

                    individual_bleu1_scores.append(bleu1)
                    individual_bleu2_scores.append(bleu2)
                    individual_bleu3_scores.append(bleu3)
                    individual_bleu4_scores.append(bleu4)
                
                # Store prediction details for CSV
                predictions_df_data.append({
                    "ImageName": img_name,
                    "GroundTruthCaptions": " || ".join(ground_truth_captions_for_image), # Join multiple GTs for display
                    "PredictedCaption": predicted_caption,
                    "BLEU-1": bleu1 if CONFIG["calculate_individual_bleu_for_histograms"] else None,
                    "BLEU-2": bleu2 if CONFIG["calculate_individual_bleu_for_histograms"] else None,
                    "BLEU-3": bleu3 if CONFIG["calculate_individual_bleu_for_histograms"] else None,
                    "BLEU-4": bleu4 if CONFIG["calculate_individual_bleu_for_histograms"] else None
                })

                # Prepare for COCO-Eval (corrected to pass list of strings directly)
                image_id = str(img_name)
                gts[image_id] = ground_truth_captions_for_image # Pass the list of strings directly
                res[image_id] = [predicted_caption] # Pass a list containing the single predicted string

                # Collect for ROUGE
                all_predictions_text.append(predicted_caption)
                all_references_text.append(ground_truth_captions_for_image)
            
            if (i+1) % CONFIG["log_step"] == 0:
                print(f"Evaluated {i+1} batches.")

    # Create and save predictions DataFrame
    predictions_df = pd.DataFrame(predictions_df_data)
    predictions_df.to_csv(CONFIG["test_predictions_file"], index=False)
    print(f"Test predictions saved to {CONFIG['test_predictions_file']}")

    # --- Calculate Evaluation Scores (COCO-Eval and ROUGE) ---
    print("\n--- Calculating Overall Evaluation Scores ---")
    scorers = [
        (Bleu(4), ["Bleu_1", "Bleu_2", "Bleu_3", "Bleu_4"]),
        # (Meteor(), "METEOR") # Uncomment if Meteor is set up
    ]

    eval_scores = {}
    for scorer, method in scorers:
        print(f"Calculating {method} score...")
        score, _ = scorer.compute_score(gts, res)
        if isinstance(method, list):
            for m, s in zip(method, score):
                eval_scores[m] = s
        else:
            eval_scores[method] = score

    # Calculate ROUGE scores for the test set
    rouge_scores_test = calculate_rouge_scores(all_predictions_text, all_references_text)
    eval_scores['ROUGE-1_P'] = rouge_scores_test['rouge1']['precision']
    eval_scores['ROUGE-1_R'] = rouge_scores_test['rouge1']['recall']
    eval_scores['ROUGE-1_F'] = rouge_scores_test['rouge1']['fmeasure']
    eval_scores['ROUGE-2_P'] = rouge_scores_test['rouge2']['precision']
    eval_scores['ROUGE-2_R'] = rouge_scores_test['rouge2']['recall']
    eval_scores['ROUGE-2_F'] = rouge_scores_test['rouge2']['fmeasure']
    eval_scores['ROUGE-L_P'] = rouge_scores_test['rougeL']['precision']
    eval_scores['ROUGE-L_R'] = rouge_scores_test['rougeL']['recall']
    eval_scores['ROUGE-L_F'] = rouge_scores_test['rougeL']['fmeasure']

    scores_df = pd.DataFrame([eval_scores])
    scores_df.to_csv(CONFIG["test_scores_file"], index=False)
    print(f"Evaluation scores saved to {CONFIG['test_scores_file']}")
    print("Overall Evaluation Scores:\n", scores_df)

    print("\n--- Evaluation Complete ---")
    return predictions_df, scores_df, individual_bleu1_scores, individual_bleu2_scores, individual_bleu3_scores, individual_bleu4_scores

# --- Custom Collate Function ---
def custom_collate_fn(batch):
    """
    Custom collate function to handle batching of tensors and lists of varying lengths.
    Ensures that all_raw_captions_list is a list of lists, matching batch size.
    """
    # Separate the items in the batch
    images = [item[0] for item in batch]
    caption_indices = [item[1] for item in batch]
    img_names = [item[2] for item in batch]
    all_raw_captions_list = [item[3] for item in batch] # This will be a list of lists

    # Stack the tensors
    images = torch.stack(images, dim=0)
    caption_indices = torch.stack(caption_indices, dim=0)

    # img_names and all_raw_captions_list are already lists of the correct batch size
    return images, caption_indices, img_names, all_raw_captions_list


# --- 8. Main Execution Block ---
def main():
    """
    Main function to run the entire image captioning pipeline:
    setup, data loading, training, and evaluation.
    """
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    os.makedirs(CONFIG["model_path"], exist_ok=True)

    feature_extractor = ViTFeatureExtractor.from_pretrained(CONFIG["vit_model_name"])
    
    image_input_size = feature_extractor.size['height'] if isinstance(feature_extractor.size, dict) else feature_extractor.size

    transform = transforms.Compose([
        transforms.Resize((image_input_size, image_input_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=feature_extractor.image_mean, std=feature_extractor.image_std)
    ])

    vocab = Vocabulary(vocab_threshold=CONFIG["vocab_threshold"], gpt2_model_name=CONFIG["gpt2_model_name"])
    
    # Store vocab and device in CONFIG for easier access in other functions
    CONFIG["vocab"] = vocab
    CONFIG["device"] = device

    dataset = Flickr8kDataset(
        image_dir=CONFIG["image_dir"],
        captions_path=CONFIG["captions_path"],
        vocab=vocab,
        transform=transform,
        max_seq_length=CONFIG["max_seq_length"],
        gpt2_model_name=CONFIG["gpt2_model_name"]
    )

    # Split the dataset into training, validation, and test sets.
    train_size = int(0.7 * len(dataset))
    val_size = int(0.15 * len(dataset))
    test_size = len(dataset) - train_size - val_size
    train_dataset, val_dataset_full, test_dataset = random_split(dataset, [train_size, val_size, test_size],
                                                            generator=torch.Generator().manual_seed(42))

    # --- Create a smaller subset for validation during training ---
    if CONFIG["val_subset_ratio"] < 1.0 and len(val_dataset_full) > 0:
        num_val_subset_samples = int(CONFIG["val_subset_ratio"] * len(val_dataset_full))
        if num_val_subset_samples == 0: # Ensure at least one sample if ratio is very small
            num_val_subset_samples = 1

        # Randomly select indices for the validation subset
        val_subset_indices = random.sample(range(len(val_dataset_full)), num_val_subset_samples)
        val_dataset_subset = Subset(val_dataset_full, val_subset_indices)
        print(f"Using a validation subset of {len(val_dataset_subset)} samples ({CONFIG['val_subset_ratio']*100:.1f}% of full validation set).")
        val_loader = DataLoader(val_dataset_subset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=8, pin_memory=True, collate_fn=custom_collate_fn)
    else:
        print("Using the full validation set.")
        val_loader = DataLoader(val_dataset_full, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=8, pin_memory=True, collate_fn=custom_collate_fn)


    train_loader = DataLoader(train_dataset, batch_size=CONFIG["batch_size"], shuffle=True, num_workers=8, pin_memory=True, collate_fn=custom_collate_fn)
    test_loader = DataLoader(test_dataset, batch_size=CONFIG["batch_size"], shuffle=False, num_workers=8, pin_memory=True, collate_fn=custom_collate_fn)

    # Train the model.
    model, train_losses, val_losses, val_bleu1_scores, val_bleu2_scores, val_bleu3_scores, val_bleu4_scores, val_meteor_scores, val_rouge1_f_scores, val_rougeL_f_scores = train_model(train_loader, val_loader, vocab, device, CONFIG)
    
    # Evaluate the trained model on the test set.
    predictions_df, scores_df, test_bleu1_scores, test_bleu2_scores, test_bleu3_scores, test_bleu4_scores = evaluate_model(model, test_loader, vocab, device, CONFIG)

    print("\nFinal Evaluation Results (Overall):")
    print(scores_df)

    # --- Plotting Training and Validation Metrics ---
    epochs_range = range(1, CONFIG["num_epochs"] + 1)

    plt.figure(figsize=(18, 6))

    # Plot Training and Validation Loss
    plt.subplot(1, 3, 1)
    plt.plot(epochs_range, train_losses, label='Training Loss')
    plt.plot(epochs_range, val_losses, label='Validation Loss (Subset)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss over Epochs')
    plt.legend()
    plt.grid(True)

    # Plot Validation BLEU Scores
    plt.subplot(1, 3, 2)
    plt.plot(epochs_range, val_bleu1_scores, label='Validation BLEU-1')
    plt.plot(epochs_range, val_bleu2_scores, label='Validation BLEU-2')
    plt.plot(epochs_range, val_bleu3_scores, label='Validation BLEU-3')
    plt.plot(epochs_range, val_bleu4_scores, label='Validation BLEU-4')
    plt.xlabel('Epoch')
    plt.ylabel('BLEU Score')
    plt.title('Validation BLEU Scores over Epochs')
    plt.legend()
    plt.grid(True)

    # Plot Validation ROUGE-F Scores
    plt.subplot(1, 3, 3)
    plt.plot(epochs_range, val_rouge1_f_scores, label='Validation ROUGE-1 F-measure', color='orange')
    plt.plot(epochs_range, val_rougeL_f_scores, label='Validation ROUGE-L F-measure', color='red')
    plt.xlabel('Epoch')
    plt.ylabel('ROUGE F-measure')
    plt.title('Validation ROUGE F-measure over Epochs')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # --- Plotting Histograms for Test Set BLEU Scores ---
    if CONFIG["calculate_individual_bleu_for_histograms"]:
        plt.figure(figsize=(12, 10))

        plt.subplot(2, 2, 1)
        plt.hist(test_bleu1_scores, bins=20, edgecolor='black')
        plt.title('BLEU-1 Distribution (Test Set)')
        plt.xlabel('BLEU-1 Score')
        plt.ylabel('Number of Samples')
        plt.grid(True, linestyle=':', alpha=0.7)

        plt.subplot(2, 2, 2)
        plt.hist(test_bleu2_scores, bins=20, edgecolor='black')
        plt.title('BLEU-2 Distribution (Test Set)')
        plt.xlabel('BLEU-2 Score')
        plt.ylabel('Number of Samples')
        plt.grid(True, linestyle=':', alpha=0.7)

        plt.subplot(2, 2, 3)
        plt.hist(test_bleu3_scores, bins=20, edgecolor='black')
        plt.title('BLEU-3 Distribution (Test Set)')
        plt.xlabel('BLEU-3 Score')
        plt.ylabel('Number of Samples')
        plt.grid(True, linestyle=':', alpha=0.7)

        plt.subplot(2, 2, 4)
        plt.hist(test_bleu4_scores, bins=20, edgecolor='black')
        plt.title('BLEU-4 Distribution (Test Set)')
        plt.xlabel('BLEU-4 Score')
        plt.ylabel('Number of Samples')
        plt.grid(True, linestyle=':', alpha=0.7)

        plt.tight_layout()
        plt.show()

    # --- Display Sample Images with Ground Truth and Predictions ---
    print("\n--- Displaying Sample Images with Captions ---")
    num_samples_to_display = 3
    
    sample_indices = random.sample(range(len(test_dataset)), min(num_samples_to_display, len(test_dataset)))

    plt.figure(figsize=(15, 7 * num_samples_to_display))

    # Collect images and predicted captions for the sample display in a batch
    sample_images_batch = []
    sample_img_names = []
    sample_ground_truth_captions_list = []

    for i, idx in enumerate(sample_indices):
        original_dataset_idx = test_dataset.indices[idx]
        image_tensor, _, original_img_name, ground_truth_captions = dataset[original_dataset_idx]
        sample_images_batch.append(image_tensor)
        sample_img_names.append(original_img_name)
        sample_ground_truth_captions_list.append(ground_truth_captions)
    
    # Stack images into a batch tensor
    predicted_captions_for_display = []
    if sample_images_batch:
        sample_images_batch_tensor = torch.stack(sample_images_batch).to(device)
        predicted_captions_for_display = model.generate_caption(sample_images_batch_tensor, vocab, device,
                                                                 max_length=CONFIG["max_seq_length"],
                                                                 beam_size=CONFIG["beam_size"])
    

    for i, idx in enumerate(sample_indices):
        original_dataset_idx = test_dataset.indices[idx]
        original_img_name = sample_img_names[i]
        ground_truth_captions = sample_ground_truth_captions_list[i]
        predicted_caption = predicted_captions_for_display[i] if i < len(predicted_captions_for_display) else "N/A"
        
        original_image_path = os.path.join(CONFIG["image_dir"], original_img_name)
        original_image = Image.open(original_image_path).convert('RGB')

        # Calculate individual BLEU scores for this sample
        references = [nltk.word_tokenize(cap.lower()) for cap in ground_truth_captions]
        candidate = nltk.word_tokenize(predicted_caption.lower())

        bleu1 = sentence_bleu(references, candidate, weights=(1, 0, 0, 0), smoothing_function=smooth_fn)
        bleu2 = sentence_bleu(references, candidate, weights=(0.5, 0.5, 0, 0), smoothing_function=smooth_fn)
        bleu3 = sentence_bleu(references, candidate, weights=(0.33, 0.33, 0.33, 0), smoothing_function=smooth_fn)
        bleu4 = sentence_bleu(references, candidate, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smooth_fn)

        plt.subplot(num_samples_to_display, 1, i + 1)
        plt.imshow(original_image)
        plt.axis('off')

        title_text = f"Image: {original_img_name}\n"
        title_text += f"Generated: {predicted_caption}\n"
        title_text += f"Ground Truth Captions:\n"
        for j, gt_cap in enumerate(ground_truth_captions):
            title_text += f"Ref {j+1}: {gt_cap}\n"
        
        title_text += f"BLEU-1 for sample: {bleu1:.4f}\n"
        title_text += f"BLEU-2 for sample: {bleu2:.4f}\n"
        title_text += f"BLEU-3 for sample: {bleu3:.4f}\n"
        title_text += f"BLEU-4 for sample: {bleu4:.4f}"
        
        plt.title(title_text, fontsize=10, loc='left', wrap=True)

    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    main()
```
